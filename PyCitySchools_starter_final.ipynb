```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete['school_name'].unique())
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete['student_name'].count()
student_count
```

```python

# Create a new DataFrame with unique 'school_name' and 'budget' values
unique_budgets = school_data_complete[['school_name', 'budget']].drop_duplicates()
# Calculate the total budget
total_budgets = unique_budgets[ 'budget'].sum()
total_budgets
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
school_data_complete.columns
# Index(['Student ID', 'student_name', 'gender', 'grade', 'school_name',
#        'reading_score', 'math_score', 'School ID', 'type', 'size', 'budget'],
#       dtype='object')
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = ({
"Total number of unique schools": school_count,
"Total students": student_count,
"Total budget": total_budgets, 
"Average math score": average_math_score, 
"Average reading score": average_reading_score,
"% passing math": passing_math_percentage,
"% passing reading": passing_reading_percentage,
"% overall passing ": overall_passing_rate   
 })


# # Display the DataFrame
district_summary_df = pd.DataFrame([district_summary])
# Formatting
district_summary_df["Total students"] = district_summary_df["Total students"].map("{:,}".format)
district_summary_df["Total budget"] = district_summary_df["Total budget"].map("${:,.2f}".format)

district_summary_df
```

## School Summary

```python
# Calculate the total student count per school
per_school_counts = school_data_complete.groupby('school_name')['Student ID'].count()
per_school_counts = pd.DataFrame(per_school_counts)
per_school_counts = per_school_counts.rename(columns={'Student ID': 'Total Students'})

per_school_counts
```

```python
per_school_budget = school_data_complete[['school_name', 'budget']].drop_duplicates()
per_school_budget.reset_index(drop = True ,inplace = True)
per_school_budget
```

```python
# Calculate the average test scores per school
# Index(['Student ID', 'student_name', 'gender', 'grade', 'school_name',
#        'reading_score', 'math_score', 'School ID', 'type', 'size', 'budget'],

per_school_math = school_data_complete.groupby('school_name')['math_score'].mean()
per_school_reading = school_data_complete.groupby('school_name')['reading_score'].mean()
```

```python
# Filter students with math scores of 70 or higher
students_passing_math = school_data_complete[school_data_complete['math_score'] >= 70]

# Group the filtered DataFrame by school_name and count the number of students in each school
school_students_passing_math = students_passing_math.groupby('school_name')['student_name'].count()
school_students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70]
school_students_passing_reading = students_passing_reading.groupby('school_name')['student_name'].count()
school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"])['student_name'].count()
school_students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates

school_summary = school_data_complete.groupby('school_name').agg({
    'school_name': 'first',
    'type': 'first',

    'student_name': 'count',
    'budget': 'first',
    'math_score': 'mean',
    'reading_score': 'mean'

})

# Rename columns for clarity
school_summary = school_summary.rename(columns={
    'school_name': 'School Name',
    'student_name': 'Total Students',
    'type': 'School Type',

    'budget': 'Total school budget',   
    'math_score': 'Average Math Score',
    'reading_score': 'Average Reading Score'
})
school_summary["Per student budget"] = (school_summary["Total school budget"] /school_summary['Total Students'])
school_summary["% passing math"] = school_students_passing_math /school_summary['Total Students']
school_summary["% passing reading"] =school_students_passing_reading /school_summary['Total Students']
school_summary["% overall passing"] = school_students_passing_math_and_reading /school_summary['Total Students']

school_summary.reset_index(drop=True, inplace = True)

new_columns = [
    'School Name',
    'School Type',

    'Total Students',
    'Total school budget',
    'Per student budget',
    'Average Math Score',
    'Average Reading Score',
    '% passing math',
    '% passing reading',
    '% overall passing'
]
school_summary = school_summary = school_summary.reindex(columns=new_columns)

# Formatting
school_summary["Total school budget"] = school_summary["Total school budget"].map("${:,.2f}".format)
school_summary["Per student budget"] = school_summary["Per student budget"].map("${:,.2f}".format)
school_summary["Average Math Score"] = school_summary["Average Math Score"].map("{:,.2f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map("{:,.2f}".format)
school_summary["% passing math"] = (school_summary["% passing math"]*100).map("{:,.2f}".format)
school_summary["% passing reading"] = (school_summary["% passing reading"]*100).map("{:,.2f}".format)
school_summary["% overall passing"] = (school_summary["% overall passing"]*100).map("{:,.2f}".format)

school_summary
```

## Highest-Performing Schools (by % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.
top_schools = school_summary.sort_values(by='% overall passing', ascending=False)
top_schools.head(5)
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = school_summary.copy()
school_spending_df 
```

```python
# Calculate the total student count per school


per_school_counts = school_data_complete.groupby('school_name')['Student ID'].count()
per_school_counts = pd.DataFrame(per_school_counts)
per_school_counts = per_school_counts.rename(columns={'Student ID': 'Total Students'})

per_school_budget = school_data_complete[['school_name', 'budget']].drop_duplicates()
per_school_budget.reset_index(drop = True ,inplace = True)

# Merge the per_school_budget and per_school_counts DataFrames on 'school_name'
school_spending= pd.merge(per_school_budget, per_school_counts, on='school_name')
school_spending['budget (Per Student)'] = school_spending['budget'] / school_spending['Total Students']

school_spending
```

```python
# Convert to float 
# school_spending['Per student budget'].astype(float) 

# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create the "Spending Ranges (Per Student)" column by binning
school_spending["Spending Ranges (Per Student)"] = pd.cut(school_spending["budget (Per Student)"], spending_bins, labels=labels)
school_spending = school_spending.sort_values('Spending Ranges (Per Student)', ascending = False)

school_spending.reset_index(drop = True, inplace = True)
school_spending.rename(columns={'school_name': 'School Name'}, inplace=True)
school_spending.sort_values(by='Spending Ranges (Per Student)')
# Scores by School Spending (5 points)
school_spending
```

```python
school_spending_df = school_spending_df.sort_values(by='Spending Ranges (Per Student)', ascending=False)
school_spending_df
```

```python
school_spending_df['Average Math Score'] = pd.to_numeric(school_spending_df['Average Math Score'], errors='coerce')
school_spending_df['Average Reading Score'] = pd.to_numeric(school_spending_df['Average Reading Score'], errors='coerce')
school_spending_df['% passing math'] = pd.to_numeric(school_spending_df['% passing math'], errors='coerce')
school_spending_df['% passing reading'] = pd.to_numeric(school_spending_df['% passing reading'], errors='coerce')
school_spending_df['% overall passing'] = pd.to_numeric(school_spending_df['% overall passing'], errors='coerce')

school_spending_df_new = school_spending_df.groupby("Spending Ranges (Per Student)").agg({
    "Average Math Score": "mean",
    "Average Reading Score": "mean",
    '% passing math': "mean",
    '% passing reading': "mean",
    '% overall passing': "mean"
}).reset_index()

school_spending_df_new
```

## Scores by School Size

```python
# school_data_complete.columns
# Index(['Student ID', 'student_name', 'gender', 'grade', 'school_name',
#        'reading_score', 'math_score', 'School ID', 'type', 'size', 'budget'],
#       dtype='object')
school_size = school_data_complete[['school_name', 'size']].drop_duplicates(subset='size')
school_size.reset_index(drop = True, inplace = True)
school_size.rename(columns =
                   {'school_name' : 'School Name' ,'size' : 'School size' 
                    }, inplace = True )
school_summary_new = school_summary.merge(school_size[['School Name', 'School size']], left_on = 'School Name',right_on='School Name')
school_summary_new

```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

# Create the "Spending Ranges (Per Student)" column by binning
school_summary_new['School size range'] = pd.cut(school_summary_new['School size'] , size_bins, labels=labels)
school_summary_new = school_summary_new.sort_values('School size range', ascending = False)
school_summary_new
```

```python
# Calculate averages for the desired columns.
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
# Display results

school_summary_new['Average Math Score'] = pd.to_numeric(school_summary_new['Average Math Score'], errors='coerce')
school_summary_new['Average Reading Score'] = pd.to_numeric(school_summary_new['Average Reading Score'], errors='coerce')
school_summary_new['% passing math'] = pd.to_numeric(school_summary_new['% passing math'], errors='coerce')
school_summary_new['% passing reading'] = pd.to_numeric(school_summary_new['% passing reading'], errors='coerce')
school_summary_new['% overall passing'] = pd.to_numeric(school_summary_new['% overall passing'], errors='coerce')
size_summary  = school_summary_new.groupby('School size range').agg({
    "Average Math Score": "mean",
    "Average Reading Score": "mean",
    '% passing math': "mean",
    '% passing reading': "mean",
    '% overall passing': "mean"
}).reset_index()
size_summary 
```
